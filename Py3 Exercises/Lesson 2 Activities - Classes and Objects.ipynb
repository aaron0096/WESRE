## Activity 4
Create a method for your `Noun` class. The method should perform a specific activity (or series of activities) on objects based on the class. 

Although this can be a simple `display`, you are encouraged to consider more complicated activities in the method.

```python
# your code here
class Coffee:
    variety = "" #Coffee variety
    origin = "" #Coffee origin
    taste = "" #Taste profile
    processing = "" #How coffee is processed
    
    def __init__(self,var,ori,t,pro):
        self.variety = var
        self.origin = ori
        self.taste = t
        self.processing = pro
    
    def display(self):
        print("This coffee type is : {}\n\
From the area of : {}\n\
And it tastes : {}\n\
And it's processed : {}\n".format(self.variety, self.origin,\
                                  self.taste,self.processing))
    
    def returninfo(self):
        return("This coffee type is : {}\n\
From the area of : {}\n\
And it tastes : {}\n\
And it's processed : {}\n".format(self.variety, self.origin,\
                                  self.taste,self.processing))
    
jairo = Coffee('Castillo','Armenia, Quindio, Colombia'\
       ,'nectarine, tangerines and kiwis, cacao undertones, strawberry liqueur finish'\
       ,'Natural ICE Fermentation')

worka = Coffee('Local Landraces and JARC 74 selections',\
                'Ethiopia - Yirgacheffe Gedeb Worka G1 Natural',\
                'Raspberries, Apricots, Wild untreated honey, Black Tea',\
                'Natural')

honduras = Coffee('Catuaí, Caturra, Pacas, Bourbon',\
                'Honduras, La Paz, Marcala',\
                'chocalty, orange, velvety taste, soft body, smooth finish and a balanced aftertaste',\
                'Washed')
    
sum_mtsinabung = Coffee('Catimor, Typica',\
                'Sumatra, Mount Sinabung',\
                'Strong Bold, Herbal, Chocolate, Floral Note, Clean and Smooth Finish',\
                'Washed')

jairo.display()
worka.display()
honduras.display()
sum_mtsinabung.display()
```

## Activity 5
Create a separate class that is logically related to your original `Noun` class, based on the example that a `Person` can have one or more `Addresses`.
	
Create at least one `Noun` object that includes a list of objects from the associated class. Display the `Noun` with the related objects.

```python
# your code here
class CoffeeShop:
    name = "" #Name of place
    addrs = "" #Place of operations
    hours = "" #time of operations
    menu = None #list of coffees for sale there
    
    def __init__(self, name, hours, addrs, menu):
        self.name = name
        self.addrs = addrs
        self.menu = menu
        self.hours = hours
    
    def details(self):
        print("{} is located at {}, and operates from {}.".format(self.name, self.addrs, self.hours))
        print("------The menu------")
        for i in self.menu:
            print(i)
    
    def add_coffee(self,coffee):
        self.menu.append(coffee)
        
    def remove_coffee(self,coffee):
        self.menu.remove(coffee)
        
coffee_list1 = []
coffee_list1.append(jairo.returninfo())

coffee_list2 = []
coffee_list2.append(worka.returninfo())

coffee_list3 = []
coffee_list3.append(honduras.returninfo())
coffee_list3.append(sum_mtsinabung.returninfo())

vdc = CoffeeShop('Venture Drive','10am-5pm','123 Bukit Merah Lane 1, #01-110',coffee_list1)
rbr = CoffeeShop('Round Boy Roasters','8am-4pm','Jalan Bukit Merah, #01-1092 Block 146, Singapore 160146',coffee_list2)
hs = CoffeeShop('HS Coffee','24hrs','Lazada',coffee_list3)

vdc.details()
rbr.details()
hs.details()

print("Add coffee ------------------------------------------------------------------------------")

vdc.add_coffee(worka.returninfo())
vdc.add_coffee(honduras.returninfo())
vdc.details()

print("Remove coffee ---------------------------------------------------------------------------")

vdc.remove_coffee(jairo.returninfo())
vdc.details()
```
